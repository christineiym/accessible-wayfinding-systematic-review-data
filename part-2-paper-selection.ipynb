## Step 0: Setup

Set up your database connection.

```python
DB_PATH: str = "./data/sqlite/analysis.db"

import sqlite3
db = sqlite3.Connection(DB_PATH)
cursor = db.cursor()
```

Check that your data has been saved as expected. We have all our tables:

```python
cursor.execute('''
               SELECT name
               FROM sqlite_master
               WHERE type='table' AND
                     name NOT LIKE 'sqlite_%' ''')
table_names = cursor.fetchall()

# List table names
print("Here are the tables in the database:")
for name in table_names:
    print(name[0])
```

And we have the expected number of records in the table of all papers.

```python
table_size: int = cursor.execute('''select count(*) from %s''' % "All_Papers").fetchone()[0]
print(table_size)
```

## Step 1: Paper Selection, Stage 1

```python
cursor.execute('''
               SELECT *
               FROM All_Papers
               LIMIT 1;''')
all_records = cursor.fetchall()
print(all_records)

# List table names
for record in all_records:
    print(record)
```

Schema info:

```python
table_info = """select name from pragma_table_info('All_Papers') as tblInfo;"""

cursor.execute(table_info)
col_names = [record[0] for record in cursor.fetchall()]

print(col_names)
```
